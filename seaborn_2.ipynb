```python
g = sns.FacetGrid(tips,col='day',row='time')
g.map(sns.violinplot,'sex','total_bill')
```

```python
g = sns.FacetGrid(tips,col='day',row='time',hue='smoker')
g.map(sns.scatterplot,'tip','total_bill')
g.add_legend()
```

```python
g = sns.FacetGrid(tips,col='day',row='time',hue='smoker')
g.map(sns.scatterplot,'sex','total_bill')
g.add_legend()
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

```python
g = sns.PairGrid(iris,hue='species')
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(iris,hue='species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.histplot)
```

```python
g = sns.PairGrid(iris,hue='species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.kdeplot)
```

```python
g = sns.PairGrid(iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
g = sns.PairGrid(iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
